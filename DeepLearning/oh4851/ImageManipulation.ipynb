```python
import os, sys
from scipy.misc import imread, imresize
import skimage.io
import skimage.transform

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

print ("packages are loaded")
```

```python
print ("test images : " + str(os.listdir('.')))
```

```python
muzi = imread('./muzi.jpg')

plt.figure(0)
plt.imshow(muzi)
plt.title('muzi! ' + str(muzi.shape))
plt.draw()
```

```python
muzi_float = imread('./muzi.jpg')

plt.figure(0)
plt.imshow(255-muzi_float)
plt.title('reversed muzi! ' + str(muzi_float.shape))
plt.draw()
```

```python
small_muzi = imresize(muzi, [200, 200, 3])

plt.figure(0)
plt.imshow(small_muzi)
plt.title('small muzi! ' + str(small_muzi.shape))
plt.draw()
```

```python
def rgb2gray(rgb_img):
    if len(rgb_img.shape) == 3:
        return np.dot(rgb_img, [0.299,0.587,0.114])
    else:
        return rgb_img
gray_muzi = rgb2gray(small_muzi)

plt.figure(0)
plt.imshow(gray_muzi, cmap=plt.get_cmap('gray'))
plt.title('gray muzi! ' + str(gray_muzi.shape))
plt.draw()
```

```python
muzi_row = np.reshape(gray_muzi, (1, -1))
print ('muzi_row shape : ' + str(muzi_row.shape))

muzi_tensor = np.reshape(gray_muzi, (200, 200))
print ('muzi_tensor shape : ' + str(muzi_tensor.shape))
```
